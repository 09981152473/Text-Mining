```python
model = gensim.models.KeyedVectors.load_word2vec_format('/content/word2vec.model-skipgram-size=200-window=5.bin',binary=True, unicode_errors='ignore')
```

```python
result = model.most_similar(positive=['پادشاه','زن'],negative=['مرد'])
```

```python
for item in result:
    print(item[0])
```

```python
# https://stackoverflow.com/questions/50275623/difference-between-most-similar-and-similar-by-vector-in-gensim-word2vec
```

```python
q = model['پادشاه'] - model['مرد'] + model['زن']
result = model.similar_by_vector(q)
for item in result:
    print(item[0])
```
